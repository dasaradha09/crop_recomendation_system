```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['label']=le.fit_transform(data['label'])
```

```python
crops={}
for i,crop in enumerate(le.classes_):
    crops[i]=crop
```

```python
import json
json.dump(crops,open('crops.json','w'))
```

```python
x=data.drop('label',axis=1)
y=data['label']
```

```python
sns.heatmap(x.corr(),annot=True,fmt=".2f")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
a=data.corr()
sns.heatmap(a,annot=True,fmt=".2f")
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
svc_model=SVC()
svc_model.fit(x_train,y_train)
y_pred_svc=svc_model.predict(x_test)
svc_score=accuracy_score(y_pred_svc,y_test)
```

```python
rf_model=RandomForestClassifier()
rf_model.fit(x_train,y_train)
y_pred_rf=rf_model.predict(x_test)
rf_score=accuracy_score(y_pred_rf,y_test)
```

```python
xgb_model=XGBClassifier()
xgb_model.fit(x_train,y_train)
y_pred_xgb=xgb_model.predict(x_test)
xgb_score=accuracy_score(y_pred_xgb,y_test)
```

```python
a=pd.DataFrame({'models':['svc','randomforest','xgboost'],'accuracy':[svc_score,rf_score,xgb_score]})
```

```python
sns.barplot(x=a['models'],y=a['accuracy'])
plt.show()
```
